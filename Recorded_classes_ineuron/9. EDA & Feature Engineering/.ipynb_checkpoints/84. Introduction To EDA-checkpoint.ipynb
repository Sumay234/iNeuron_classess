```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

## Display all columns of the dataset--
pd.pandas.set_option("display.max_columns",None)
```

```python
dataset=pd.read_csv("house_prediction_train.csv")
```

```python
## shape of dataset with rows and columns--

dataset.shape
```

## In Data analysis we will analyze to find the below stuff

1. Missing Values
2. All the Numerical Variable
3. Distribution Of the numerical variable
4. Categorical Variable
5. Outliers
6. Relationship between independence and dependence variable
7. Correlation

```python
## 1. Missing Values--
dataset.isnull().sum()
```

```python
## only that columns which have null value--

features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>=1]
features_with_na
```

```python
## to get percentage of null columns--

for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean()*100,4),"% missing value")
```

```python
sns.boxplot(dataset["SalePrice"])
```

Observation:

Here are many Outlier in the SalePrice so we will use Median instead of mean

```python
for feature in features_with_na:
    data=dataset.copy()
    # lets make a variable that indicate 1 for missing value & 0 for rest
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    # lets calculate the mean SalePrice where the information is missing
    # here we are using median as we are asumming that there might be outlier
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.title(feature)
    plt.show()
```

observation:
          
    Here with the relation between the missing values and the dependent variable is clearly visible. So, we need to replace the nan value with something meaningfull which we will do in the Feature Engineering section.

## Numerical Variable

```python
## To get only Object Columns from Dataset--

numerical_features=[]
for feature in dataset.columns:
    if dataset[feature].dtypes != "O":
        numerical_features.append(feature)
        
numerical_features
```

```python
[feature for feature in dataset.columns if dataset[feature].dtype != "O"]
```

```python
len(numerical_features)
```

```python
dataset[numerical_features].head(5)
```

```python
## Temporal Variable(Eg- DateTime Variable)


for year_feature in numerical_features:
    if "Yr" in year_feature or "Year" in year_feature:
        print(year_feature)
```

```python
year_feature=[yr_feature for yr_feature in numerical_features if 'Yr' in yr_feature or "Year" in yr_feature]
year_feature
```

```python
for unique_feature in year_feature:
    print(unique_feature, dataset[unique_feature].unique())
```

```python
dataset.groupby("YrSold")["SalePrice"].median().plot()
plt.title("House Price Vs Sale Price")
```

```python
## Here we will compare the difference of year feature with sale feature--

data=dataset.copy()
for feature in year_feature:
    if feature != "YrSold":
        data[feature] = data["YrSold"] - data[feature]
        
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title(feature)
        plt.show()
```

Observation:

    Within all the yearwise information, it is clearly visible that the the New Houses are more costlier than the old ones. Looks like a pareto or power law distribution.

```python
## numerical_variable--
## 1) Continuous variable & 2) Discrete Variable

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique()) <=25]
print(len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
data=dataset.copy()

for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot()
    plt.show()
```

```python
data=dataset.copy()

for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

observation:
    
    There is  a relationship between discrete feature and saleprice 
    as it increasing the price is also increasing 

```python
## Continuous Variable--

continuous_feature=[]
for feature in numerical_features:
    if feature not in discrete_feature+year_feature:
        continuous_feature.append(feature)
        
continuous_feature
```

```python
len(continuous_feature)
```

```python
## Lets analysis the continuous feature with the histogram---

data=dataset.copy()
for feature in continuous_feature:
    data.groupby(feature)["SalePrice"].median().plot.hist()
    plt.xlabel(feature)
    plt.title(feature)
    plt.show()
```

```python
data=dataset.copy()
for feature in continuous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.title(feature)
    plt.show()
```

## EDA Part-2

```python
data=dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title(feature)
        plt.show()
```

```python
## Outliers--

for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
