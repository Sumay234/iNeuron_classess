## Pruning  -- ccp

```python
path=dt_model_1.cost_complexity_pruning_path(x_train,y_train)
```

```python
path
```

```python
ccp_alphas=path.ccp_alphas
ccp_alphas
```

```python
dt_model_2=[]
for ccp in ccp_alphas:
    dt_m =DecisionTreeClassifier(ccp_alpha=ccp)
    dt_m.fit(x_train,y_train)
    dt_model_2.append(dt_m)
```

```python
dt_model_2
```

```python
train_score=[i.score(x1,y1) for i in dt_model_2]
train_score
```

```python
test_score=[i.score(x_test,y_test) for i in dt_model_2]
test_score
```

```python
plt.figure(figsize=(20,20))
fig,ax=plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for traing and testing alpha")
ax.plot(ccp_alphas,train_score,drawstyle='steps-post',label='train',marker='o')
ax.plot(ccp_alphas,test_score,drawstyle='steps-post',label='test',marker='o')

ax.legend()
plt.show()
```
